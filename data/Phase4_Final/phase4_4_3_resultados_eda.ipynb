```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import os
import json
from scipy.stats import mannwhitneyu

# Style and Seed
plt.style.use('seaborn-v0_8-whitegrid')
np.random.seed(42)

# Paths
FIGURES_DIR = "phase4_figures"
TABLES_DIR = "phase4_tables"
os.makedirs(FIGURES_DIR, exist_ok=True)
os.makedirs(TABLES_DIR, exist_ok=True)

# Inputs
# PHASE3_CSV = "../../data/Phase3_Final/phase3_finalresults.csv" # Replaced by generator
PHASE3_CSV = 'C:/Users/emore/TFG/LISBETH-main/LISBETH-main/data/Phase3_Final/phase3_finalresults.csv'
PHASE2_OCCURRENCES = "../../data/phase2/embeddings_occurrences.csv" # If exists
# DIMENSIONS_JSON = "C:/Users/emore/TFG/LISBETH-main/LISBETH-main/data/dimensiones_ancla_conflicto.json" # Replaced by generator
DIMENSIONS_JSON = ''

print("Setup Complete. SEED=42.")
```

```python

# Load Data
try:
    df = pd.read_csv(PHASE3_CSV)
    df['window_end_month'] = pd.to_datetime(df['window_end_month'])
    df.sort_values('window_end_month', inplace=True)
    df.set_index('window_end_month', inplace=True)
    print(f"Loaded Phase 3 Results: {len(df)} rows")
except Exception as e:
    print(f"CRITICAL FAIL: Could not load {PHASE3_CSV}")
    raise e

# Helpers
variants = ["baseline"]
if any("dapt" in c for c in df.columns):
    variants.append("dapt")
strategies = ["penultimate", "last4_concat"]
dims = ["violence_security", "humanitarian_civilian", "legitimacy_responsibility"]

print(f"Variants: {variants}")
print(f"Strategies: {strategies}")
```

```python
import matplotlib as mpl
import matplotlib.dates as mdates

mpl.rcParams['font.family'] = 'DejaVu Sans'
df.index = pd.to_datetime(df.index)

variant_label_en = {
    "baseline": "Baseline",
    "dapt": "DAPT"
}
strategy_label_en = {
    "penultimate": "Penultimate (penultimate layer)",
    "last4_concat": "Last4Concat (concat. last 4 layers)"
}

# A. Drift by strategy and variant
for s in strategies:
    plt.figure(figsize=(12, 6))

    for v in variants:
        col = f"drift_{v}_{s}"
        if col in df.columns:
            v_lbl = variant_label_en.get(v, v)
            plt.plot(df.index, df[col], label=v_lbl, marker='.')

    s_lbl = strategy_label_en.get(s, s)
    plt.title(f"Drift between subspaces (Grassmann) across time\nStrategy: {s_lbl}")
    plt.xlabel("Temporal window")
    plt.ylabel("Drift (Grassmann)")
    plt.grid(True, alpha=0.3)
    plt.legend(title="Variant", frameon=False)

    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.savefig(f"{FIGURES_DIR}/43_drift_timeline_{s}.png", bbox_inches='tight')
    plt.show()

# B. Deriva anotada (Baseline + Penultimate)
target_col = "drift_baseline_penultimate"

if target_col in df.columns:
    top_5 = df[target_col].nlargest(5)

    plt.figure(figsize=(15, 7))
    plt.plot(
        df.index, df[target_col],
        label="Baseline · Penultimate",
        marker='o'
    )

    # Annotate top 5 peaks
    for date, val in top_5.items():
        plt.annotate(
            f"{date.strftime('%Y-%m')}\n{val:.3f}",
            (date, val),
            xytext=(0, 12),
            textcoords='offset points',
            ha='center',
            arrowprops=dict(arrowstyle="->")
        )

    plt.title("Drift (Grassmann) with highlighted peaks (Top 5)\nBaseline · Penultimate")
    plt.xlabel("Temporal window")
    plt.ylabel("Drift (Grassmann)")
    plt.grid(True, alpha=0.3)
    plt.legend(frameon=False)

    # 🔥 CLAVE
    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.savefig(f"{FIGURES_DIR}/43_drift_timeline_annotated.png", bbox_inches='tight')
    plt.show()
# C. Tabla de rankings (sin cambios, no es gráfico)
rankings_list = []
for v in variants:
    for s in strategies:
        col = f"drift_{v}_{s}"
        if col in df.columns:
            # Top Drift
            top_d = df[col].nlargest(5)
            for d, val in top_d.items():
                rankings_list.append({
                    "metric": "drift", "variant": v, "strategy": s,
                    "window": d, "value": val, "type": "absolute"
                })

            # Top Delta Drift
            delta = df[col].diff().abs()
            top_delt = delta.nlargest(5)
            for d, val in top_delt.items():
                rankings_list.append({
                    "metric": "drift", "variant": v, "strategy": s,
                    "window": d, "value": val, "type": "delta"
                })

pd.DataFrame(rankings_list).to_csv(f"{TABLES_DIR}/43_key_windows_rankings.csv")
print("Exported rankings.")
```

```python
# --- 2. Proyecciones Semánticas (Heatmaps - Comparativa A/B) ---

import matplotlib as mpl
mpl.rcParams['font.family'] = 'DejaVu Sans'  # buena cobertura y menos warnings
df.index = pd.to_datetime(df.index)

extrema_list = []
conditions = ["raw", "corrected"]

# Etiquetas solo para visualización (no afectan a columnas/archivos)
variant_label_es = {
    "baseline": "Baseline",
    "dapt": "DAPT"
}
strategy_label_es = {
    "penultimate": "Penultimate (penultimate layer)",
    "last4_concat": "Last4Concat (concat. last 4  layer)"
}
condition_label_es = {
    "raw": "No correction",
    "corrected": "Corrected"
}

# A. SUBSPACE PROJECTION (métrica principal: alineación de la “forma” semántica)
print("\n=== GENERATING THE SUBSPACES ALIGNMENT HEATMAPS ===")
for cond in conditions:
    suffix = f"_{cond}"

    for s in strategies:
        for v in variants:
            cols = [f"subspace_proj_{d}_{v}_{s}{suffix}" for d in dims]

            if all(c in df.columns for c in cols):
                data = df[cols].copy()
                data.columns = dims
                data.index = data.index.strftime('%Y-%m')

                plt.figure(figsize=(12, 6))
                ax = sns.heatmap(
                    data.T,
                    cmap="viridis",
                    annot=False,
                    vmin=0, vmax=1,
                    cbar=True,
                    cbar_kws={"label": "Subspace alignment (0–1)"}
                )

                v_lbl = variant_label_es.get(v, v)
                s_lbl = strategy_label_es.get(s, s)
                c_lbl = condition_label_es.get(cond, cond)

                plt.title(
                    "Alignment of semantic subspaces across time and dimensions",
                    fontsize=14,
                    fontweight='bold',
                    pad=15
                )

                plt.suptitle(
                    f"Variant: {v_lbl} · Strategy: {s_lbl} · Condition: {c_lbl}",
                    fontsize=10,
                    y=0.92
                )

                plt.xlabel("Time window", fontsize=11)
                plt.ylabel("Semantic dimension", fontsize=11)

                ax.tick_params(axis='x', labelsize=10)
                ax.tick_params(axis='y', labelsize=10)

                plt.tight_layout(rect=[0, 0, 1, 0.9])
                plt.savefig(f"{FIGURES_DIR}/43_heatmap_subspace_{s}_{v}_{cond}.png")
                plt.show()

                # Extrema Tracking (Subspace)
                for d in dims:
                    c_name = f"subspace_proj_{d}_{v}_{s}{suffix}"
                    if df[c_name].isna().all():
                        continue
                    idmax = df[c_name].idxmax()
                    if pd.isna(idmax):
                        continue

                    extrema_list.append({
                        "type": "subspace", "dim": d, "variant": v, "strategy": s,
                        "condition": cond,
                        "window": idmax, "val": df.loc[idmax, c_name], "extum": "max"
                    })

# B. CENTROID PROJECTION (métrica secundaria: posición del “centro” semántico)
print("\n=== GENERATING HEATMAPS OF CENTROID PROJECTIONS ===")
for cond in conditions:
    suffix = f"_{cond}"

    for s in strategies:
        for v in variants:
            cols = [f"centroid_proj_{d}_{v}_{s}{suffix}" for d in dims]

            if all(c in df.columns for c in cols):
                data = df[cols].copy()
                data.columns = dims
                data.index = data.index.strftime('%Y-%m')

                plt.figure(figsize=(12, 6))
                ax = sns.heatmap(
                    data.T,
                    cmap="RdBu_r",
                    center=0,
                    annot=False,
                    cbar=True,
                    cbar_kws={"label": "Centroid projection (signed value)"}
                )

                v_lbl = variant_label_es.get(v, v)
                s_lbl = strategy_label_es.get(s, s)
                c_lbl = condition_label_es.get(cond, cond)

                plt.title(
                    "Semantic centroid projection across time and dimensions",
                    fontsize=14,
                    fontweight='bold',
                    pad=15
                )

                plt.suptitle(
                    f"Variant: {v_lbl} · Strategy: {s_lbl} · Condition: {c_lbl}",
                    fontsize=10,
                    y=0.92
                )
                plt.xlabel("Time window", fontsize=11)
                plt.ylabel("Semantic dimension", fontsize=11)

                ax.tick_params(axis='x', labelsize=10)
                ax.tick_params(axis='y', labelsize=10)

                plt.tight_layout(rect=[0, 0, 1, 0.9])
                plt.savefig(f"{FIGURES_DIR}/43_heatmap_centroid_{s}_{v}_{cond}.png")
                plt.show()

pd.DataFrame(extrema_list).to_csv(f"{TABLES_DIR}/43_projection_extrema_dual.csv", index=False)
```

```python
# --- 3. Sensibilidad y Densidad ---

import matplotlib as mpl
mpl.rcParams['font.family'] = 'DejaVu Sans'  # buena cobertura de caracteres
df.index = pd.to_datetime(df.index)

# A. Métricas vs n_occurrences
metrics = ['drift', 'entropy']
target_s = 'penultimate'
target_v = 'baseline'

metric_label_es = {
    "drift": "Drift between subspaces (Grassmann)",
    "entropy": "Spectral entropy"
}
variant_label_es = {"baseline": "Baseline", "dapt": "DAPT"}
strategy_label_es = {
    "penultimate": "Penultimate (penultimate layer)",
    "last4_concat": "Last4Concat (concat. last 4 layer)"
}

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharex=True)

for i, m in enumerate(metrics):
    col = f"{m}_{target_v}_{target_s}"
    if col in df.columns:
        sns.scatterplot(
            data=df,
            x='n_occurrences',
            y=col,
            ax=axes[i],
            alpha=0.6
        )

        m_lbl = metric_label_es.get(m, m)
        v_lbl = variant_label_es.get(target_v, target_v)
        s_lbl = strategy_label_es.get(target_s, target_s)

        axes[i].set_title(f"{m_lbl} vs number of occurrences\nVariant: {v_lbl} · Strategy: {s_lbl}")
        axes[i].set_xlabel("Number of occurences in the window (n_occurrences)")
        axes[i].set_ylabel(m_lbl)
        axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/43_metric_vs_nocc.png")
plt.show()

# B. Efecto de baja densidad
# Comparar distribuciones para ventanas de baja densidad vs normales
if 'low_density' in df.columns:
    from scipy.stats import mannwhitneyu

    ld_windows = df[df['low_density'] == True]
    normal_windows = df[df['low_density'] == False]

    sens_summary = []

    metric_cols = [c for c in df.columns if 'drift' in c or 'entropy' in c]
    # Filtrar para estrategia y variante objetivo
    metric_cols = [c for c in metric_cols if target_s in c and target_v in c]

    for col in metric_cols:
        stat, p = mannwhitneyu(
            ld_windows[col].dropna(),
            normal_windows[col].dropna()
        )
        sens_summary.append({
            "metric": col,
            "mean_low_density": ld_windows[col].mean(),
            "mean_normal": normal_windows[col].mean(),
            "mann_whitney_p": p,
            "significant_diff": p < 0.05
        })

    pd.DataFrame(sens_summary).to_csv(f"{TABLES_DIR}/43_sensitivity_summary.csv", index=False)

    # Plot comparativo (boxplot)
    if metric_cols:
        plt.figure(figsize=(12, 6))

        plot_df = df.melt(id_vars=['low_density'], value_vars=metric_cols)

        # Etiquetas de métrica solo para el gráfico (sin tocar nombres de columnas)
        def pretty_metric_name(colname: str) -> str:
            if "drift" in colname:
                return "Drift (Grassmann)"
            if "entropy" in colname:
                return "Spectral Entropy"
            return colname

        plot_df["metric"] = plot_df["variable"].apply(pretty_metric_name)
        plot_df["density"] = plot_df["low_density"].map({
            True: "Low density",
            False: "Normal density"
        })

        sns.boxplot(
            data=plot_df,
            x="metric",
            y="value",
            hue="density"
        )

        v_lbl = variant_label_es.get(target_v, target_v)
        s_lbl = strategy_label_es.get(target_s, target_s)

        plt.title(
            "Effect of low density on metrics\n"
            f"Variant: {v_lbl} · Strategy: {s_lbl}"
        )
        plt.xlabel("Metric")
        plt.ylabel("Metric value")
        plt.grid(True, axis='y', alpha=0.3)
        plt.legend(title="Tipe of window", frameon=False)

        plt.tight_layout()
        plt.savefig(f"{FIGURES_DIR}/43_low_density_effect.png")
        plt.show()
```
